# Временной шаг для заданного числа Куранта в зависимости от скорости и шага по пространству

Co = dt * |U| / dx

Фиксируем число Куранта, варьируем U и dx

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
def calc_dt(dx, U, Co):
    '''Вычисление шага по времени в зависимости от числа Куранта'''
    return Co*dx / U
```

```python
dx = [i*0.001 for i in range(1, 30)]
U = [i*20 for i in range(1, 30)]
dt = []
```

```python
Co = 1.  # Задаёмся числом Куранта
for i in range(len(U)):
    for j in range(len(dx)):
        dt.append(calc_dt(dx[j], U[i], Co))
```

```python
x, y = np.meshgrid(dx, U)
z = np.array(dt).reshape(len(dx), len(U))
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z, color='red')
ax.set_xlabel('dx, м')
ax.set_ylabel('U, м/c')
ax.set_zlabel('dt, c')
s = 'Число Куранта = ' + str(Co)
ax.set_title(s);
```
